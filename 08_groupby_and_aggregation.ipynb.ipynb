```python
import pandas as pd
df = pd.read_csv('/Users/kenantufankurt/Desktop/PYTHON/PANDAS/COREY SCHAFER/stack-overflow-developer-survey-2019/survey_results_public.csv')
```

```python
country_group = df.groupby('Country')
```

```python
country_group.get_group('United States')
```

```python
df.groupby('Country').get_group('United States')
```

```python
df.groupby('Country').get_group('India')
```

```python
filt = df['Country'] == 'United States'
df.loc[filt]
```

```python
filt = df['Country'] == 'United States'
df.loc[filt]['SocialMedia'].value_counts()
```

```python
country_group = df.groupby('Country')
country_group['SocialMedia'].value_counts(normalize = True).head(50)
```

```python
country_group['ConvertedComp'].agg(['median' , 'mean']).loc['Albania']
```

```python
filt = df['Country'] == 'United States'
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
filt = df['Country'] == 'United States'
df.loc[filt]['LanguageWorkedWith'].apply(lambda x: x == x.str.contains('Python').sum())
```
